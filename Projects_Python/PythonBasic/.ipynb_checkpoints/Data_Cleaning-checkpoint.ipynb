```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\thean\Downloads\Customer Call List.xlsx')
df
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
df.drop(columns = 'Not_Useful_Column', inplace = True)
df
```

```python
#C1 df['Last_Name'] = df['Last_Name'].str.strip('/_.')
df['Last_Name'] = df['Last_Name'].str.replace(r'[/._]', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex = True)
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[nNan-]', '', regex = True)
df
```

```python
df[['Street Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand = True)
df
```

```python
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace({'Y': 'Yes', 'N': 'No'})
df = df.fillna('N/a')
df
```

```python
for x in df.index:
    temp = df.loc[x, 'Do_Not_Contact']
    if temp == 'Yes' or temp == 'N/a':
        df.drop(x, inplace = True)
df
#df.dropna(subnet='Phone_Number', inplace = True) 
```
